```python
import pandas as pd 
df = pd.read_csv('Combined_dataset.csv')
df.head(100)
```

```python
print('Shape:',df.shape)
print('Columns:',df.columns.tolist())
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
for col  in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].fillna('Unknown')

for col in df.select_dtypes(exclude=['object']).columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
df = df.drop_duplicates()
```

```python
filtered_df = df[df['rating'] > 4]
filtered_df[['title','rating']].head()
```

```python
df['quantity'] = 1
df['final_price_num'] = pd.to_numeric(df['final_price'].astype(str).str.extract(r'(\d+\.?\d*)')[0], errors='coerce').fillna(0)
df['total_amount'] = df['final_price_num'] * df['quantity']
df[['final_price_num','quantity','total_amount']].head()
```

```python
df.to_csv('cleaned_dataset.csv', index=False)
```
